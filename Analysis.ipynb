```python
%%sql 
select count(*) from film f inner join inventory i on (f.film_id = i.film_id)
    inner join rental r on (i.inventory_id = r.inventory_id)
                    inner join payment p on (r.customer_id = p.customer_id )
               
                   
```

```python
%%sql
SELECT count(*)
FROM payment p
inner JOIN rental r    ON ( p.rental_id = r.rental_id )
inner JOIN inventory i ON ( r.inventory_id = i.inventory_id )
inner JOIN film f      ON ( i.film_id = f.film_id)
```

```python
%%sql 
select p.customer_id, p.rental_id, p.amount, ci.city  from payment p 
            join customer c on (p.customer_id= c.customer_id)
            join address a on (a.address_id = c.address_id)
            join city ci on (ci.city_id = a.city_id)
            order by p.payment_date
            limit 10
```

```python
%%sql
SELECT p.customer_id, p.rental_id, p.amount, ci.city                            
FROM payment p
JOIN customer c  ON ( p.customer_id = c.customer_id )
JOIN address a ON ( c.address_id = a.address_id )
JOIN city ci ON ( a.city_id = ci.city_id )
order by p.payment_date
limit 10;
```

```python
%%sql
select c.city, sum(p.amount) as revenue from city c 
        join address a on (a.city_id = c.city_id)
        join customer cu on (a.address_id = cu.address_id)
        join payment p on (p.customer_id = cu.customer_id)
        GROUP BY c.city
        ORDER BY revenue DESC
        limit 10
        
```

```python
%%sql
SELECT ci.city, sum(p.amount) as revenue
FROM payment p
JOIN customer c  ON ( p.customer_id = c.customer_id )
JOIN address a ON ( c.address_id = a.address_id )
JOIN city ci ON ( a.city_id = ci.city_id )
GROUP BY ci.city
ORDER BY revenue DESC
limit 10
```
